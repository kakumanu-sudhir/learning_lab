**Creating multi-dimensional arrays**

np.array creates an array object

Lets Create a 1-D array

```
one_d_array = np.array([1,2,3])
print(one_d_array)
print(one_d_array.shape)
```

set the x co-ordinates with one_d_array and y coordinate to 1 for all the x-values, ie. points become (1,1), (2,1) & (3,1)

```
plt.plot(one_d_array, len(one_d_array) * [1], "o")
plt.show()
```

Creation of a 2-D array

```
two_d_array = np.array([[1,2],[4,5], [6,7]])
print(two_d_array)
print(two_d_array.shape)
```

To visualize lets get the x value of each coordinate into one list and y into another.

We can transpose the matrix to get 2 dimensions to plot

```
print(two_d_array.T)
print(two_d_array.T.shape)
plt.plot(two_d_array.T[0], two_d_array.T[1], "o")
plt.show()
```

You can see that the points are at (1,2) , (4,5), (6,7)

## Widely used **operations** 

1-D Convolution:

Consider two vectors [5,2,1,0,6,3] and [-1, 0, 1]

Lets apply **Convolve** operation on that:

*   Convolve operation is a general purpose filter effect for images.
*   It is done by multiplying a pixel’s and its neighboring pixels color value by a matrix, i.e. dot product of the input data layer with a filter matrix

For the computation, the conv_filter i.e. kernel would be reversed (for cross-corelation this vector doesnt get reversed)


          5 2 1 0 6 3
              
      1 0 -1                   5*-1 = -5
        1 0 -1                 5*0 + 2*-1 = -2
          1 0 -1               5*1 + 2*0 + 1*-1 = 4 
            1 0 -1             2*1 + 1*0 + 0*-1 = 2
              1 0 -1           1*1 + 0*0 + 6*-1 = -5
                1 0 -1         1*0 + 6*0 + -1*3 = -3
                  1 0 -1       1*6 + 3*0 = 6
                    1 0 -1     1*3 = 3


## Convolve using Numpy

```
input_layer = np.array([5,2,1,0,6,3])
conv_filter = np.array([-1,0,1])
stride = 1

#The output is the same size as in1, centered with respect to the ‘full discrete linear convolution of the inputs’.
print(np.convolve(input_layer, conv_filter, 'same'))

#The output consists only of those elements that do not rely on the zero-padding, i.e. Take the inputs ignoring the padding.
print(np.convolve(input_layer, conv_filter, 'valid'))
```

# Visual representation of differences between Convolution and Cross-corelation operations.



*   Convolution is a mathematical operation on two functions (f and g) that produces a third function expressing how the shape of one is modified by the other.
*   Cross-correlation is a measure of similarity of two series as a function of the displacement of one relative to the other. This is also known as a sliding dot product or sliding inner-product
*   Autocorrelation, which is the cross-correlation of a signal with itself, there will always be a peak at a lag of zero, and its size will be the signal energy.
